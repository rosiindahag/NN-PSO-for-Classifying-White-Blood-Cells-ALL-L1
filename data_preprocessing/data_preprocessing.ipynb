```python
#Read dataset file
data = pd.read_csv('ekstraksi.csv')
```

```python
#nambah identitas output1
output1 = [] #1 = lymphoid (lymphoblast and lymphocyte), 0 = non-lymphoid, for first stage NN-PSO
output2 = [] #1 = lymphoblast, 0 = non-lymphoblast (lymphocyte), 2 = non-lymphoid, for multiclass classification
output3 = [] #1 = lymphoblast, 0 = non-lymphoblast (lymphocyte), for second stage NN-PSO
X = np.array(data["Citra"]) #original name label
#print(X)
for i in range(len(data)):
    if X[i] == "Limfoblas":
        output1.append(1)
        output2.append(1)
        output3.append(1)
    elif X[i] == "Limfosit":
        output1.append(1)
        output2.append(0)
        output3.append(0)
    elif X[i] == "Limfoma":
        output1.append(1)
        output2.append(0)
        output3.append(0)
    else:
        output1.append(0)
        output2.append(2)
        output3.append(0)
y = output1 #label output1
y2 = output2 #label output2
y3 = output3 #label output3
data = data.drop(['Citra'], axis=1) #delete the name label
```

```python
#normalisasi
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
len_databaru = np.size(data, axis=1)-1
data.iloc[:,:len_databaru] = scaler.fit_transform(data.iloc[:,:len_databaru])
data
```
